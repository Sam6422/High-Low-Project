```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

import time
from IPython.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
fb = pd.read_excel(r"NSE_Stock_List.xlsx")
b = fb['stock']
```

```python
dates = ['2023-05-16', '2023-05-17', '2023-05-18', '2023-05-19', '2023-05-22', '2023-05-23', '2023-05-24', 
    '2023-05-25', '2023-05-26', '2023-05-29', '2023-05-30', '2023-05-31', '2023-06-01', '2023-06-02', 
    '2023-06-05', '2023-06-06', '2023-06-07', '2023-06-08', '2023-06-09', '2023-06-12', '2023-06-13', '2023-06-14',
    '2023-06-15', '2023-06-16', '2023-06-19', '2023-06-20', '2023-06-21']

# lists required
min9_15am = 0
min10_15am = 0
min11_15am = 0
min12_15pm = 0
min1_15pm = 0
min2_15pm = 0
min3_15pm = 0

max9_15am = 0
max10_15am = 0
max11_15am = 0
max12_15pm = 0
max1_15pm = 0
max2_15pm = 0
max3_15pm = 0
```

```python
for ticker in range(28, 1500):    ## date-wise
    for date in range(0,len(dates)-1):      ## stock-wise
        try:
            data = yf.download(tickers= b[ticker], start=dates[date], end=dates[date+1],interval="1h").reset_index()
            data['Date'] = data['Datetime'].dt.date
            data['Time'] = data['Datetime'].dt.time
            data = data.loc[data.index <= 6, :]
            data = data.dropna(how='all')

            min_row = data['Low'].idxmin()
            max_row = data['High'].idxmax()

            if data.iloc[min_row,8] == datetime.time(9, 15):
                min9_15am += 1
            elif data.iloc[min_row,8] == datetime.time(10, 15):
                min10_15am += 1
            elif data.iloc[min_row,8] == datetime.time(11, 15):
                min11_15am += 1
            elif data.iloc[min_row,8] == datetime.time(12, 15):
                min12_15pm += 1
            elif data.iloc[min_row,8] == datetime.time(13, 15):
                min1_15pm += 1
            elif data.iloc[min_row,8] == datetime.time(14, 15):
                min2_15pm += 1
            elif data.iloc[min_row,8] == datetime.time(15, 15):
                min3_15pm += 1

            if data.iloc[max_row,8] == datetime.time(9, 15):
                max9_15am += 1
            elif data.iloc[max_row,8] == datetime.time(10, 15):
                max10_15am += 1
            elif data.iloc[max_row,8] == datetime.time(11, 15):
                max11_15am += 1
            elif data.iloc[max_row,8] == datetime.time(12, 15):
                max12_15pm += 1
            elif data.iloc[max_row,8] == datetime.time(13, 15):
                max1_15pm += 1
            elif data.iloc[max_row,8] == datetime.time(14, 15):
                max2_15pm += 1
            elif data.iloc[max_row,8] == datetime.time(15, 15):
                max3_15pm += 1
        
        except:
            pass
    print(f"{ticker}--{b[ticker]}")
```

```python
c = ['9-10am', '10-11am', '11-12am', '12-1pm', '1-2pm', '2-3pm', '3-4pm']
a = [min9_15am, min10_15am, min11_15am, min12_15pm, min1_15pm, min2_15pm, min3_15pm]
b = [max9_15am, max10_15am, max11_15am, max12_15pm, max1_15pm, max2_15pm, max3_15pm]
mintime_maxtime = pd.DataFrame({'timeframe':c, 'min': a, 'max': b})
mintime_maxtime
```

# Plotting the results

```python
sns.displot(mintime_maxtime, x='max', kind='hist')
```

```python
plt.plot(mintime_maxtime['timeframe'], mintime_maxtime['min'], lw=2, color='red', label='price at lows')
plt.plot(mintime_maxtime['timeframe'], mintime_maxtime['max'], lw=2, color='green', label='price at highs')
plt.xlabel('timeframe', fontsize=20)
plt.ylabel('frequency', fontsize=20)
# plt.ylim(-5,25)
plt.legend()
plt.show()
```
